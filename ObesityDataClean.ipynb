After downloading this clean CSV, I shortened Time Period cells by changing 2015/16 to 2015, 2016/17 to 2016 etc.

I looked up any boundary name changes that have occured between 2013 and 2021, and changed the names of 2013 data in my GeoJSON coordinates file (via Visual Studio Code) so that they would match up in the labelled LA map, and so that I could show the most recent data. I did this using this spreadsheet from ONS (https://github.com/jesme00/jesme00.github.io/blob/main/LA%20Name%20Changes.xlsx) as well as Google searches.  

For example, what historically was split into two districts, 'West Somerset' and 'Taunton Deane', is now a merged area called 'Somerset West'.

This was a better method than using 2020/21 boundaryies in the GeoJSON from the offset, which is a much larger file size than the 2013 boundary one. Since there have been only very minor physical boundary changes, these changes mostly involved renaming a couple of 2013 districts next to each other with the same new merged 2020/21 UA name. This is sufficent when mapping country-wide data because the areas are visually very small. I may have wanted to be more pedantic with changing physical boundaries if I was zooming into a particular part of England.
